```python
import os
import cv2
import numpy as np
#from skimage.metrics import structural_similarity as ssim

def calculate_metrics(faces_path, srcnn_path):
    normal_image = cv2.imread(faces_path)
    srcnn_image = cv2.imread(srcnn_path)
    srcnn_image = cv2.resize(srcnn_image, (normal_image.shape[1], normal_image.shape[0]))
    normal_image_gray = cv2.cvtColor(normal_image, cv2.COLOR_BGR2GRAY)
    srcnn_image_gray = cv2.cvtColor(srcnn_image, cv2.COLOR_BGR2GRAY)
    psnr_value = cv2.PSNR(normal_image_gray, srcnn_image_gray)
    return psnr_value

input = 'face_112_79.jpg'
srcnn = 'enhanced_face_112_79.jpg'
calculate_metrics(input , srcnn)
```
